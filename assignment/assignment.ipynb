```python
# Q4

import seaborn as sns
sns.histplot(df['grad_100_value'])
```

```python
sns.kdeplot(df['grad_100_value'])
```

```python
sns.boxplot(df['grad_100_value'])
```

```python
df['grad_100_value'].describe()
```

```python
# 5

sns.kdeplot(data=df,x='grad_100_value',hue='control')
df.loc[:,['control','grad_100_value']].groupby('control').describe()
```

```python
sns.kdeplot(data=df,x='grad_100_value',hue='level')
df.loc[:,['level','grad_100_value']].groupby('level').describe()
```

The average graduation rates for private not-for-profit schools is 42% compared to 29% for private for-profit and 17% for public colleges. 4-year institutions on average have a higher graduation rate of 33% compared to 22% for 2-year institutions.

```python
# Q6

df['levelXcontrol']=df['level']+', '+df['control']
sns.kdeplot(data=df,x='grad_100_value',hue='levelXcontrol')
```

```python
df.loc[:,['level','control','grad_100_value']].groupby(['level','control']).describe()
```

Private not-for-profit schools, both 2-year and 4-year, have the best graduation rates. Their rates are about the same and nearly twice the amount of the private for-profit 2-year colleges. The difference seems to lie more in the control than with the level.
